## Task #1: Rename Multiple Columns in a list

```python
df = df.rename(columns = {'Player': 'player', 'Span': 'span', 'Ave': 'ave', 'Runs': 'runs', 'Mat': 'matches', 'NO': 'not_outs', 'HS': 'highest_inns_score', 'BF': 'balls_faced', 'SR': 'batting_strike_rate'})
```

```python
df.head(5)
```

## Task #2: Check Null Values

```python
df.isnull().any()
```

```python
df[df['balls_faced'].isna() == 1]
```

```python
df['balls_faced'] = df['balls_faced'].fillna(0)
df['batting_strike_rate'] = df['batting_strike_rate'].fillna(0)
```

```python
df[df['player'] == 'CL Walcott (WI)']
```

## Task #3: Drop duplicates

```python
df.duplicated()
```

```python
df[df['player'].duplicated() == 1]
```

## Task #4: Split up 'span' into 'start_date' and 'end_date'

```python
df['rookie_year'] = df['span'].str.split(pat = '-').str[0]
```

```python
df['final_year'] = df['span'].str.split(pat = '-').str[1]
```

```python
df.head(5)
```

## Task #5: Drop 'span' column

```python
df = df.drop(['span'], axis = 1)
```

```python
df.head(5)
```

## Task 6: Split up  the 'country' from the player

```python
df['country'] = df['player'].str.split(pat = '(').str[1].str.split(pat = ')').str[0]
```

```python
df['player'] = df['player'].str.split(pat = '(').str[0]
```

```python
df.head(5)
```

## Task #7: Change DataTypes

```python
# highest_inns_score -> int
# balls_faces -> int
# rookie_year -> int
# final_year -> int
# 0 -> int
# 4s -> int
# 6s -> int 

df.dtypes
```

```python
df['highest_inns_score'] = df['highest_inns_score'].str.split(pat = '*').str[0]
```

```python
df['highest_inns_score'] = df['highest_inns_score'].astype('int')
```

```python
df = df.astype({'rookie_year': 'int', 'final_year': 'int'})
```

```python
df['balls_faced'] = df['balls_faced'].str.split(pat = '+').str[0]
```

```python
df['balls_faced'] = df['balls_faced'].fillna(0)
```

```python
df['balls_faced'] = df['balls_faced'].astype('int')
```

```python
df['0'] = df['0'].fillna(0)
df['0'] = df['0'].astype('int')
```

```python
df['4s'] = df['4s'].str.split(pat = '+').str[0]
df['4s'] = df['4s'].astype('int')
```

```python
df['6s'] = df['6s'].str.split(pat = '+').str[0]
df['6s'] = df['6s'].astype('int')
```

```python
df.dtypes
```

## Task #8: Build out 'career_length' column

```python
df['career_length'] = df['final_year'] - df['rookie_year']
```

```python
df.head(5)
```

## Question #1: What is the Average Career Length?

```python
df['career_length'].mean()
```

## Question #2: AVG batting strike rate for cricketers who played over 10 years

```python
df[df['career_length'] > 10]['batting_strike_rate'].mean()
```

## Question #3: Find number of cricketers who played before 1960

```python
df[df['rookie_year'] < 1960]['player'].count()
```

## Question #4: Max highest inns score by country

```python
df.groupby('country')['highest_inns_score'].max().to_frame('high_inn_score').reset_index().sort_values('high_inn_score', ascending = False)
```
